## Data Cleaning

```python
#Checking for duplicate values:

dup_data = raw_data[raw_data.duplicated()]
dup_data.shape
```

```python
total = pd.isnull(raw_data).sum().sort_values(ascending = False)
percentage = ((pd.isnull(raw_data).sum() / pd.isnull(raw_data).count()).sort_values(ascending=False))*100
no_values = pd.concat([total,percentage],axis = 1, keys = ['Total', 'Percent'])
```

```python
no_values.head(19)
```

```python
raw_data_without_null = raw_data.drop((no_values[no_values['Total']>1]).index,1)
```

```python
raw_data_without_null = raw_data_without_null.dropna()
```

```python
pd.isnull(raw_data_without_null).sum().max()
```

## Checking conditions for Linear Model

```python
data_cleaned = raw_data_without_null.copy()
```

```python
data_use = data_cleaned[factors].copy()
```

```python
data_use['LogPrice'] = np.log(data_use['SalePrice'])
```

```python
location = data_cleaned[['Condition1', 'Neighborhood']].copy()
```

```python
data = data_cleaned[factors]
data_with_cat = data.join(location)
```

```python
data_with_cat
```

```python
data_with_cat['LogPrice'] = np.log(data_with_cat['SalePrice'])
data_with_cat =data_with_cat.drop(['SalePrice'], axis = 1)
```

```python
data_with_dummies = pd.get_dummies(data_with_cat, drop_first = True)
```

```python
data_with_dummies
```

```python
data_with_dummies.columns
```

```python
cols = ['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF',
       '1stFlrSF', 'FullBath', 'TotRmsAbvGrd', 'YearBuilt', 'LogPrice',
       'Condition1_Feedr', 'Condition1_Norm', 'Condition1_PosA',
       'Condition1_PosN', 'Condition1_RRAe', 'Condition1_RRAn',
       'Condition1_RRNe', 'Condition1_RRNn', 'Neighborhood_Blueste',
       'Neighborhood_BrDale', 'Neighborhood_BrkSide', 'Neighborhood_ClearCr',
       'Neighborhood_CollgCr', 'Neighborhood_Crawfor', 'Neighborhood_Edwards',
       'Neighborhood_Gilbert', 'Neighborhood_IDOTRR', 'Neighborhood_MeadowV',
       'Neighborhood_Mitchel', 'Neighborhood_NAmes', 'Neighborhood_NPkVill',
       'Neighborhood_NWAmes', 'Neighborhood_NoRidge', 'Neighborhood_NridgHt',
       'Neighborhood_OldTown', 'Neighborhood_SWISU', 'Neighborhood_Sawyer',
       'Neighborhood_SawyerW', 'Neighborhood_Somerst', 'Neighborhood_StoneBr',
       'Neighborhood_Timber', 'Neighborhood_Veenker']
```

```python
inputs = data_with_dummies.drop(['LogPrice'],axis =1)
```

```python
target = data_with_dummies['LogPrice']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
scaler = StandardScaler()
scaler.fit(inputs)
scaled_inputs = scaler.transform(inputs)
```

```python
X_train, X_test, y_train,y_test = train_test_split(scaled_inputs, target, test_size = 0.2, random_state= 365)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
y_pred = linear_model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_pred, y_test)
mae
```
